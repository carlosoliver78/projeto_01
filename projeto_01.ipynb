### 1) Os três grupos de depressão diferem quanto à idade?

Teste ANOVA (F): compara as médias de mais de dois grupos

Hipóteses do teste ANOVA:

● H0: não há diferença entre as médias dos grupos

● H1: há pelo menos uma diferença entre as médias dos grupos

```python
from scipy.stats import f_oneway
```

```python
df_aux = tabela3[["phq9", "idade"]].dropna()


f_oneway(df_aux[df_aux["phq9"] == 0]["idade"]
         , df_aux[df_aux["phq9"] == 1]["idade"]
         , df_aux[df_aux["phq9"] == 2]["idade"])  
```

Como o p-valor é menor que 0.5, logo, rejeitamos a HO. Verificamos que pelo menos um grupo tem médias diferentes.

### 2) Os três grupos de depressão diferem quanto ao tempo dedicado a exercício físico, por semana?
Teste ANOVA (F): compara as médias de mais de dois grupos

Hipóteses do teste ANOVA:

● H0: os três grupos de depressão não diferem quanto à
média de tempo gasto com exercício físico

● Ha: há pelo menos uma diferença na média de tempo gasto
com exercício físico entre os três grupos de depressão   

```python
df_aux = tabela3[["phq9", "tempo_atividade_semanal_pag"]].dropna()

f_oneway(df_aux[df_aux["phq9"] == 0]["tempo_atividade_semanal_pag"]
         , df_aux[df_aux["phq9"] == 1]["tempo_atividade_semanal_pag"]
         , df_aux[df_aux["phq9"] == 2]["tempo_atividade_semanal_pag"])
```

pvalor < 0.05, portanto, rejeitamos H0

Há pelo menos uma diferença na média de tempo gasto com exercício físico entre os três grupos de depressão.

Quais grupos diferem? Vamos usar um Post Hoc
Para responder a essa pergunta, faremos o teste post-hoc de Tukey

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(df_aux["tempo_atividade_semanal_pag"], df_aux["phq9"], alpha = 0.05)

print(tukey)

# 0.0    489.060397
# 1.0    455.335846
# 2.0    286.915637
```

Quando p-adj menor que 0.05, rejeitamos H0 (par tem a mesma média).

Dessa forma, rejeitamos o H0 para os pares 0-2 e o 1-2 (médias não são iguais)

### 3) A proporção de pessoas em cada grupo de depressão varia de acordo com o gênero?

Teste qui-quadrado: avalia a associação entre duas variáveis categóricas

Hipóteses do teste qui-quadrado:
    
● H0: não há associação entre as duas variáveis
    
● Ha: há associação entre as duas variáveis

```python
from scipy.stats import chi2_contingency

crosstab = pd.crosstab(tabela3["genero"], tabela3["phq9"])
crosstab 
```

```python
chi2_contingency(crosstab)  
```
